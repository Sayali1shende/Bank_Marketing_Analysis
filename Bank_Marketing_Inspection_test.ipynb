# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
from google.colab import files


# 2. Upload file
uploaded = files.upload()


```

```python


# Load CSV
df = pd.read_csv("bankmarketing.csv")
print(df.head())


```

```python
# load dataset
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:\n", df.head())

```

```python
# 3. Check Datatypes & Missing Values

print("\nData types of each column:\n", df.dtypes)
print("\nMissing values in each column:\n", df.isnull().sum())

total_missing = df.isnull().sum().sum()
print("\nTotal missing values:", total_missing)
if total_missing == 0:
    print("✅ No missing values in dataset.")
else:
    print("⚠️ Missing values exist, need to handle them.")

```

```python
# 4. Summary Statistics

print("\nSummary statistics (numeric features):\n")
print(df.describe())
```

```python
# 5. Target Distribution

target_col = "y"   # target variable
print("\nTarget value counts:\n", df[target_col].value_counts())

sns.countplot(x=df[target_col])
plt.title("Target Distribution (Subscribed to Term Deposit)")
plt.show()
```

```python
# 6. Correlation Heatmap (numeric features)

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
# 7. Additional Insights

# Distribution of age
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Subscription rate by job
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="job", hue="y")
plt.title("Subscription Rate by Job Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Subscription rate by education
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="education", hue="y")
plt.title("Subscription Rate by Education Level")
plt.xticks(rotation=45)
plt.show()
```
